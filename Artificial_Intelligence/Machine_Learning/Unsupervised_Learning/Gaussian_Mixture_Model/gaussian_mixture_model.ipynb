# Gaussian Mixture Model - Mô hình hỗn hợp Gaussian

## 1. Giới thiệu

Gaussian Mixture Model (GMM) là một mô hình xác suất được dùng trong học không giám sát để phân cụm dữ liệu. GMM giả định rằng dữ liệu được tạo thành từ sự kết hợp của nhiều phân phối chuẩn (Gaussian distributions), mỗi phân phối đại diện cho một cụm.

Khác với K-Means chỉ gán mỗi điểm dữ liệu vào một cụm duy nhất, GMM tính xác suất một điểm thuộc về mỗi cụm, cho phép phân cụm "mềm" (soft clustering).

## 2. Mô hình xác suất

GMM biểu diễn hàm mật độ xác suất tổng quát như sau:

$$
p(x) = \sum_{i=1}^{k} \pi_i \cdot \mathcal{N}(x | \mu_i, \Sigma_i)
$$

- $k$: số cụm (thành phần Gaussian)
- $\pi_i$: trọng số (mixing coefficient) của cụm thứ $i$, với $\sum_{i=1}^{k} \pi_i = 1$
- $\mathcal{N}(x | \mu_i, \Sigma_i)$: phân phối chuẩn đa biến có trung tâm $\mu_i$ và ma trận hiệp phương sai $\Sigma_i$

## 3. Phân cụm mềm (Soft Clustering)

Với mỗi điểm $x$, GMM tính xác suất nó thuộc về cụm $i$:

$$
\gamma_{i}(x) = \frac{\pi_i \cdot \mathcal{N}(x | \mu_i, \Sigma_i)}{\sum_{j=1}^{k} \pi_j \cdot \mathcal{N}(x | \mu_j, \Sigma_j)}
$$

- $\gamma_{i}(x)$: xác suất (responsibility) điểm $x$ thuộc về cụm $i$

## 4. Huấn luyện mô hình: EM Algorithm

GMM thường được huấn luyện bằng thuật toán **Expectation-Maximization (EM)** gồm hai bước lặp:

1. **E-step**: Tính xác suất mỗi điểm thuộc về từng cụm.
2. **M-step**: Cập nhật các tham số $\pi_i$, $\mu_i$, $\Sigma_i$ dựa trên các xác suất vừa tính.

Lặp lại hai bước trên cho đến khi hội tụ.

## 5. Ưu điểm và Nhược điểm

### ✅ Ưu điểm:
- Mô hình linh hoạt, biểu diễn được các cụm có hình dạng phức tạp.
- Phân cụm mềm, cho kết quả mượt hơn so với K-Means.

### ❌ Nhược điểm:
- Nhạy cảm với khởi tạo và nhiễu.
- Dễ rơi vào cực trị cục bộ.
- Phức tạp tính toán hơn K-Means.

## 6. Ứng dụng thực tế

- Nhận dạng giọng nói và âm thanh.
- Phân tích ảnh và xử lý hình ảnh.
- Mô hình hóa xác suất trong dữ liệu tài chính.
- Phân loại văn bản và trích xuất chủ đề.
