# K-Means Clustering - Phân cụm K-Means

## 1. Giới thiệu

K-Means là một thuật toán học không giám sát (Unsupervised Learning) phổ biến dùng để phân cụm dữ liệu. Mục tiêu của K-Means là chia dữ liệu thành $k$ cụm sao cho các điểm trong cùng một cụm càng giống nhau càng tốt, còn khác cụm thì càng khác biệt.

## 2. Cách hoạt động của K-Means

1. Chọn số cụm $k$.
2. Khởi tạo $k$ tâm cụm (centroids) ngẫu nhiên.
3. Gán mỗi điểm dữ liệu vào cụm có tâm gần nhất.
4. Cập nhật lại tâm cụm bằng cách tính trung bình các điểm trong cụm.
5. Lặp lại bước 3 và 4 cho đến khi:
   - Các cụm không thay đổi nữa, hoặc
   - Đạt số lần lặp tối đa.

## 3. Công thức và khái niệm quan trọng

### 3.1 Hàm mục tiêu (Objective Function)

K-Means tối ưu hàm tổng bình phương khoảng cách từ mỗi điểm đến tâm cụm của nó:

$$
J = \sum_{i=1}^{k} \sum_{x \in C_i} \| x - \mu_i \|^2
$$

- $k$: số cụm
- $C_i$: cụm thứ $i$
- $\mu_i$: tâm cụm $i$
- $\| x - \mu_i \|$: khoảng cách Euclidean từ điểm $x$ đến tâm cụm

### 3.2 Khoảng cách Euclidean

Khoảng cách phổ biến nhất trong K-Means:

$$
d(x, y) = \sqrt{(x_1 - y_1)^2 + (x_2 - y_2)^2 + \cdots + (x_n - y_n)^2}
$$

## 4. Cách chọn số cụm $k$

### 4.1 Elbow Method

- Vẽ biểu đồ giữa số cụm $k$ và tổng sai số bình phương (SSE - Sum of Squared Errors).
- Điểm “khuỷu tay” (elbow) là nơi nên chọn $k$ vì sau đó SSE giảm không đáng kể.

### 4.2 Silhouette Score

Đo độ “tốt” của việc phân cụm:

$$
s = \frac{b - a}{\max(a, b)}
$$

- $a$: khoảng cách trung bình từ một điểm đến các điểm khác trong cùng cụm.
- $b$: khoảng cách trung bình nhỏ nhất đến các điểm trong cụm gần nhất khác.

## 5. Ưu điểm và Nhược điểm

### ✅ Ưu điểm:
- Dễ hiểu, dễ triển khai.
- Tính toán nhanh với dữ liệu lớn.

### ❌ Nhược điểm:
- Cần chọn trước số cụm $k$.
- Nhạy cảm với tâm cụm khởi tạo.
- Không hoạt động tốt với cụm có hình dạng phức tạp hoặc kích thước khác nhau.

## 6. Ứng dụng thực tế

- Phân nhóm khách hàng (customer segmentation).
- Nhận dạng ảnh, nén ảnh.
- Phát hiện bất thường.
- Tìm chủ đề trong văn bản (topic modeling).
