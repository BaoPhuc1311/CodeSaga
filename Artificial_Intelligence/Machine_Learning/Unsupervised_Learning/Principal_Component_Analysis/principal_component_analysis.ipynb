# Principal Component Analysis (PCA) - Phân tích thành phần chính

## 1. Giới thiệu
PCA (Principal Component Analysis) là một kỹ thuật giảm chiều dữ liệu (dimensionality reduction) được sử dụng phổ biến trong học máy và khai phá dữ liệu. Mục tiêu chính của PCA là tìm ra những thành phần chính (principal components) giúp giữ lại phần lớn thông tin của dữ liệu ban đầu nhưng với ít chiều hơn.

## 2. Cách hoạt động của PCA

1. **Chuẩn hóa dữ liệu**:
   - Trừ trung bình và chia cho độ lệch chuẩn để mỗi thuộc tính có giá trị trung bình bằng 0 và phương sai bằng 1.

2. **Tính ma trận hiệp phương sai (Covariance Matrix)**:
   $$
   \text{Cov}(X) = \frac{1}{n-1} X^T X
   $$
   - Trong đó \( X \) là ma trận dữ liệu đã được chuẩn hóa.

3. **Tính toán giá trị riêng (Eigenvalues) và vector riêng (Eigenvectors)**:
   - Các vector riêng tương ứng với các thành phần chính.
   - Các giá trị riêng thể hiện lượng phương sai được giải thích bởi từng thành phần chính.

4. **Chọn số lượng thành phần chính**:
   - Thông thường chọn \( k \) thành phần chính đầu tiên sao cho tổng phương sai giải thích được là đủ lớn (ví dụ 95%).

5. **Chiếu dữ liệu lên không gian mới**:
   $$
   X_{new} = X \cdot W_k
   $$
   - \( W_k \) là ma trận chứa \( k \) vector riêng ứng với \( k \) giá trị riêng lớn nhất.

## 3. Các công thức quan trọng

### 3.1 Phương sai (Variance)
$$
\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2
$$

### 3.2 Ma trận hiệp phương sai
$$
\text{Cov}(X, Y) = \frac{1}{n - 1} \sum_{i=1}^{n}(X_i - \bar{X})(Y_i - \bar{Y})
$$

### 3.3 Tổng phương sai giải thích (Explained Variance Ratio)
$$
\text{EVR}_k = \frac{\lambda_k}{\sum_{j=1}^{n} \lambda_j}
$$
- \( \lambda_k \): giá trị riêng thứ \( k \).
- Dùng để xác định phần trăm phương sai mà thành phần chính thứ \( k \) giải thích.

## 4. Ứng dụng của PCA

- **Giảm chiều dữ liệu**: Chuẩn bị dữ liệu cho mô hình học máy bằng cách giảm số lượng đặc trưng đầu vào.
- **Thị giác máy tính (Computer Vision)**: PCA được dùng trong nhận diện khuôn mặt, nén ảnh.
- **Tiền xử lý dữ liệu**: Khử nhiễu, chuẩn hóa không gian đặc trưng.
- **Trực quan hóa dữ liệu**: Chiếu dữ liệu nhiều chiều xuống 2D hoặc 3D để trực quan hóa.
